### Import the required libraries

```python
import os.path as op
import mne
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mne.stats import permutation_cluster_test
from nilearn.plotting import plot_connectome

# in order to have interactive plots please uncomment the following line:
# %matplotlib qt
```

### Load data

```python
conds = ['wake', 'nrem']
kinds = ['hdeeg', 'seeg']

data = {k: {c: [] for c in conds} for k in kinds }
for k in kinds:    
    for c in conds:
        fname = './data/s1_%s_%s-raw.fif' % (c, k)
        data[k][c] = mne.io.read_raw_fif(fname)
```

### Load SEEG spatial information and 3d surfaces

```python
# SEEG channel information
fname_seeg_ch_info = './data/s1_seeg_ch_info.csv'
seeg_ch_info = pd.read_csv(fname_seeg_ch_info)

surf_kinds = ['outer_skin', 'brain']
surfs = {k: mne.read_surface('./data/s1_%s' % k) for k in surf_kinds}
```

### Find events
We will extract the events corresponding to the electrical stimulations

```python
events = {k: {c: [] for c in conds} for k in kinds}
for k in kinds:
    for c in conds:
        events[k][c] = mne.find_events(data[k][c])
```

### Plot timecourses
We will have a look at the timecourses:

```python
for k in kinds:
    _ = data[k]['wake'].plot(events=events[k]['wake'])
```

### Spatial information

Here we can the seeg channels' information as the coordinates in native MRI space and MNI space and the correspondig anatomical area of each contact.


```python
seeg_ch_info.head()
```

Let's see the spatial distribution of the seeg channels:

```python
seeg_ch_info = seeg_ch_info.loc[seeg_ch_info.name.isin(data['seeg']['wake'].ch_names)] # keep only recorded channels
coords = seeg_ch_info[['x_norm_mri', 'y_norm_mri', 'z_norm_mri']].values # get MNI coordinates
adj_mat = np.zeros((len(coords), len(coords)))
plot_connectome(adj_mat, node_coords=coords, node_color='tab:green',
                                        display_mode='ortho', node_size=15, black_bg=True)
```

Now let's see it in 3D:

```python
import mayavi.mlab as mlab
s_coords = seeg_ch_info[['x_surf', 'y_surf', 'z_surf']].values # get MNI coordinates

for t in surf_kinds:
    mlab.triangular_mesh(surfs[t][0][:, 0], surfs[t][0][:, 1], surfs[t][0][:, 2], surfs[t][1], opacity=0.3, color=(0.7, 0.7, 0.7))
    mlab.points3d(s_coords[:, 0], s_coords[:, 1], s_coords[:, 2], scale_mode='vector', scale_factor=3,colormap='bwr')
    
                                         
mlab.show()
```

Now let's see the EEG channel locations:

```python
_ = data['hdeeg']['wake'].plot_sensors()
```

## Extract epochs

Here we extract the time locked responses to intracranial stimulation

```python
epochs = {k: {c: [] for c in conds} for k in kinds }
for k in kinds:
    for c in conds:
        epochs[k][c] = mne.Epochs(data[k][c], events[k][c],tmin=-0.25, tmax=0.5, baseline=(None, -0.05))
```

```python
for k in kinds:
    _ = epochs[k]['wake'].plot()
```

### Plot the evoked response
And look at the evoked responses in a butterfly plot:

```python
for c in conds:
    print(c)
    _ = epochs['hdeeg'][c].average().plot(spatial_colors=True)
    _ = epochs['seeg'][c].average().plot()
```

### Comparison
Get the data from one channel on each recording type and condition:

```python
chans = {'hdeeg': ['e61'], 'seeg': ['E10']}
cond_data = {k: {c: [] for c in conds} for k in kinds }

for k in kinds:
    for c in conds:
        cond_data[k][c] = epochs[k][c].get_data()[:, mne.pick_channels(epochs[k][c].ch_names, chans[k]), :].squeeze()

```

### Plot comparison between conditions

```python
for k in kinds:
    plt.figure()
    for c in conds:
        plt.plot(epochs[k][c].times, cond_data[k][c].mean(0))
        plt.fill_between(epochs[k][c].times, cond_data[k][c].mean(0)- cond_data[k][c].std(0), 
                         cond_data[k][c].mean(0)+ cond_data[k][c].std(0), alpha = 0.3)
    
    plt.vlines(0, ymin=plt.gca().get_ylim()[0], ymax=plt.gca().get_ylim()[1], linestyles='--', alpha=0.5)
    plt.hlines(0, xmin=plt.gca().get_xlim()[0], xmax=plt.gca().get_xlim()[1], linestyles='--', alpha=0.5)
    plt.legend(['wake', 'nrem'])
    plt.ylabel('Amplitude (V)')
    plt.xlabel('Time (s)')
    plt.title('%s - Channel: %s' % (k, chans[k][0]))


```
